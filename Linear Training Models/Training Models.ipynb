*So far we have applied the closed form model for linear regression, just the normal equation.. let us try gradient descent where we iteratively reach the optimal solution in a few steps*

#### Looking at a basic implementation of Batch Gradient Descent ####

```python
eta = 0.2 ## learning rate
n_epochs = 10 ## no of iterations
m = len(X_b) 
plt.plot(X,y,"b.") 
plt.grid()

theta = np.random.randn(2,1) ## random initialisation of theta

for epoch in range(n_epochs):
    gradient = 2/m * X_b.T @ (X_b @ theta - y)
    theta = theta - eta* gradient
    predicted = X_b @ theta
    plt.plot(X, predicted, 'r--')
```

```python
theta
```

*We can see the red line (our prediction) slowly converging to the optimal solution... even after a random intialisation of theta*

### Stochastic Gradient Descent ###

```python
n_epochs = 1
t0, t1 = 5, 50 # learning schedule hyperparameters
def learning_schedule(t):
    return t0 / (t + t1) ## We wish to slowly decrease the learning rate as the no of iteration grows (we arrive near the globa optima)
    
np.random.seed(42)
theta = np.random.randn(2,1) # random initialization
plt.plot(X,y,"b.")
plt.grid()
for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        predicted = X_b @ theta
        plt.plot(X, predicted, 'r--')
```

```python
theta
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-5, penalty=None, eta0=0.01,
                       n_iter_no_change=100, random_state=42)
sgd_reg.fit(X, y.ravel())  # y.ravel() because fit() expects 1D targets
```

```python
sgd_reg.intercept_, sgd_reg.coef_
```

*The mini batch gradient descent is one which solves the problem of neither being too random (like stochastic) or too slow (like batch). It trains on smaller batches of the training set in any given iteration*

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")

lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
plt.grid()

```

```python
from sklearn.model_selection import learning_curve
train_sizes, train_scores, valid_scores = learning_curve(LinearRegression(), X, y, train_sizes=np.linspace(0.01,1.0,40), cv =5, scoring = "neg_root_mean_squared_error")
train_errors = -train_scores.mean(axis=1)
valid_errors = -valid_scores.mean(axis=1)

plt.plot(train_sizes, train_errors, 'r--', linewidth = 2, label= 'train')
plt.plot(train_sizes, valid_errors, 'b--', linewidth = 2, label= 'valid')
plt.grid()
```


*The above model is underfitting so far -> both the errors meet at a point and after that they just **plateau***

trying a 5 degree polynomial model

```python
from sklearn.pipeline import Pipeline

polynomial_reg=  Pipeline([
    ("poly_features", PolynomialFeatures(degree=5, include_bias=False)),
    ('lin_reg', LinearRegression()),
])

plot_learning_curves(polynomial_reg, X, y)
```

**That gap between training and validation error indicates some overfitting**

A good question is how would our optimal model look like.. let's test it out and see

```python
from sklearn.pipeline import Pipeline

polynomial_reg=  Pipeline([
    ("poly_features", PolynomialFeatures(degree=2, include_bias=False)), ## we have used a quadratic eq to generate the data
    ('lin_reg', LinearRegression()),
])

plot_learning_curves(polynomial_reg, X, y)
```

***This** where both the **training error** and the **validation error** are really low and they meet before plateauing together is indicative of the right model*

### Early Stopping

*We wish to stop our learning algo as soon as we feel we have reached our minima, continuing might cause overfit (especially with a high learning rate)*

```python
from copy import deepcopy 
from sklearn.metrics import root_mean_squared_error
from sklearn.preprocessing import StandardScaler

def early_stopping(X, y):
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2)
    
    preprocessing= Pipeline([
        ("poly_90", PolynomialFeatures(degree=90, include_bias= False)),
        ("scaler", StandardScaler())
    ])
    
    X_train_prep= preprocessing.fit_transform(X_train)
    X_valid_prep= preprocessing.transform(X_valid)
    
    sgd_reg = SGDRegressor(penalty= None, eta0= 0.002, random_state= 42)
    n_epochs= 500
    best_valid_rmse= float('inf')
    
    for epoch in range(n_epochs):
        sgd_reg.partial_fit(X_train_prep, y_train.ravel())
        y_valid_predict = sgd_reg.predict(X_valid_prep)
        val_error= root_mean_squared_error(y_valid, y_valid_predict)
    
        if val_error < best_valid_rmse: 
            best_valid_rmse= val_error
            best_model = deepcopy(sgd_reg)
```

```python
test_ratio = 0.2
validation_ratio = 0.2
total_size = len(X_with_bias)

test_size = int(total_size * test_ratio)
validation_size = int(total_size * validation_ratio)
train_size = total_size - test_size - validation_size


rnd_indices = np.random.permutation(total_size)
```

```python
rnd_indices
```

```python
X_train = X_with_bias[rnd_indices[:train_size]]
y_train = y[rnd_indices[:train_size]]
X_valid = X_with_bias[rnd_indices[train_size:-test_size]]
y_valid = y[rnd_indices[train_size:-test_size]]
X_test = X_with_bias[rnd_indices[-test_size:]]
y_test = y[rnd_indices[-test_size:]]
```

```python
mean = X_train[:, 1:].mean(axis=0)
std = X_train[:, 1:].std(axis=0)
X_train[:, 1:] = (X_train[:, 1:] - mean) / std
X_valid[:, 1:] = (X_valid[:, 1:] - mean) / std
X_test[:, 1:] = (X_test[:, 1:] - mean) / std
```

```python
def to_one_hot(y):
    return np.diag(np.ones(y.max() + 1))[y]
```

```python
Y_train_one_hot = to_one_hot(y_train)
Y_valid_one_hot = to_one_hot(y_valid)
Y_test_one_hot = to_one_hot(y_test)
```

```python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = exps.sum(axis=1, keepdims=True)
    return exps / exp_sums
```

```python
n_inputs = X_train.shape[1]  # == 3 (2 features plus the bias term)
n_outputs = len(np.unique(y_train))  # == 3 (there are 3 iris classes)
```

```python
eta = 0.5
n_epochs = 10000
m = len(X_train)

Theta = np.random.randn(n_inputs, n_outputs) ## Each class has it's own 3 parameters.. bias, sepal length and width
print("Validation Losses: \n")

for epoch in range(n_epochs):
    logits = X_train @ Theta
    Y_proba = softmax(logits)
    if epoch % 1000 == 0:
        Y_proba_valid = softmax(X_valid @ Theta)
        xentropy_losses = -(Y_valid_one_hot * np.log(Y_proba_valid)) 
        print(epoch, xentropy_losses.sum(axis=1).mean()) ## sum of y(i)* log(p(i))
    error = Y_proba - Y_train_one_hot
    gradients = 1 / m * X_train.T @ error
    Theta = Theta - eta * gradients
```
