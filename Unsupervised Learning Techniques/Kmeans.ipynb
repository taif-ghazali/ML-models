### Clustering

*It is the task of **identifying similar instances** and assigning them to **clusters**, or groups of similar instances*

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, y= make_blobs(
    n_samples=300,  # Total number of points
    centers= 5,  # Number of clusters (blobs)
    n_features=2,  # Number of dimensions/columns
    cluster_std=1.0,  # Spread/standard deviation of clusters
    random_state=42,  # Seed for reproducible results])
)

k=5
kmeans= KMeans(n_clusters= k, random_state= 42)
y_pred= kmeans.fit_predict(X)
```

```python
import numpy as np


def plot_boundary(clf, X):
    
    x_min, x_max= X[:, 0].min() -0.5, X[:, 1].max() + 0.5
    y_min, y_max= X[:, 1].min() -0.5, X[:, 1].max() + 0.5

    
    y_pred= clf.fit_predict(X)
    
    xx, yy= np.meshgrid(
        np.linspace(x_min, x_max, 100), 
        np.linspace(y_min, y_max, 100)
    )
    X_new= np.c_[xx.ravel(), yy.ravel()]
    zz= clf.predict(X_new).reshape(xx.shape)
    
    plt.figure(figsize= (10, 4))
    
    plt.contourf(xx, yy, zz, alpha= 0.3, cmap= "tab10")
    plt.scatter(X[:,0], X[:,1], c= y_pred, cmap= "tab10", s= 25)
    
    plt.scatter(
        clf.cluster_centers_[:, 0], 
        clf.cluster_centers_[:, 1], 
        marker = "X", 
        s= 200, 
        c= "black",
        label= "Centroids"
    )
    
    plt.grid()
    plt.legend()
    plt.show()

plot_boundary(kmeans, X)
```

### Image Segmentation

```python
import PIL
from IPython.display import display

img= PIL.Image.open("pearl_poster.jpg").resize((500,650))
image= np.asarray(img)
display(img)
```

So let's have that as our original image.. have a look at this...

```python
image
```

```python
image.shape
```

*So that's our image array.. it's of **3 dimensions** height, width and then **3 channels for (R, G, B) combo**.. it's actually pretty cool if you think how these values actually combine together to give the image.. just a smart comboo of RGB values*

Now, see here..

```python
X= image.reshape(-1,3)
kmeans = KMeans(n_clusters= 8, random_state= 42).fit(X)
segmented_img= kmeans.cluster_centers_[kmeans.labels_]
segmented_img= segmented_img.reshape(image.shape)
```

```python
segmented_img.shape
```

```python
segmented_img.dtype
```

```python
fig, axes= plt.subplots(1,2, figsize= (12,8))

axes[0].imshow(np.array(img))
axes[1].imshow(segmented_img.astype(np.uint8))
```

You can see clearly, even with only 8 clusters, we did a pretty dope job recovering that image

```python
fig, axes= plt.subplots(1,4, figsize= (12,8))
X= image.reshape(-1,3)
i=0

for k in np.arange(2,9,2):
    kmeans = KMeans(n_clusters= k).fit(X)
    segmented_img= kmeans.cluster_centers_[kmeans.labels_]
    segmented_img= segmented_img.reshape(image.shape)
    axes[i++].imshow(segmented_img.astype(np.uint8))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,12))
X = image.reshape(-1, 3)

for ax, k in zip(axes.ravel(), range(2, 10, 2)):   # k = 2,4,6,8
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X)
    segmented_img = kmeans.cluster_centers_[kmeans.labels_]
    segmented_img = segmented_img.reshape(image.shape)

    ax.imshow(segmented_img.astype(np.uint8))
    ax.set_title(f"k = {k}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

### Clustering in Semi Supervised Learning

```python
from sklearn.datasets import load_digits

X_digits, y_digits= load_digits(return_X_y= True)
X_train, y_train= X_digits[:1400], y_digits[:1400]
X_test, y_test = X_digits[1400: ], y_digits[1400: ]
```

Say we only have labels for 50 instances

```python
from sklearn.linear_model import LogisticRegression

n_labeled= 50
log_reg= LogisticRegression(max_iter= 10_000)
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
```

```python
log_reg.score(X_test, y_test)
```

```python
X_train[:5]
```

```python
y_train[:5]
```

```python
plt.imshow(X_train[1].reshape(8,8), cmap="Greys")
```

*So our basic logistic regression scores only a 75% accuracy on this set (We ony trained on the first 50 images.. hence **semi supervised**)*

```python
k= 50
kmeans= KMeans(n_clusters= k, random_state= 42)
X_digits_dist= kmeans.fit_transform(X_train)
representative_digit_idx= np.argmin(X_digits_dist, axis= 0)
```

```python
representative_digit_idx
```

```python
X_representative_digits= X_train[representative_digit_idx]
```

```python
X_representative_digits.shape
```

```python
fig, axes= plt.subplots(5, 10, figsize= (12,2))

for ax, i in zip(axes.ravel(), range(0,50)):
    ax.imshow(X_representative_digits[i].reshape(8,8), cmap="Greys")
    ax.axis("off")
```

```python
y_representative_digits= np.array([8,0,1,3,6,7,5,4,2,8,2,3,9,5,3,9,1,7,9,1,4,6,9,7,5,2,2,1,3,3,6,0,4,9,8,1,8,4,2,4,2,3,9,7,8,9,6,5,6,4])
```

```python
y_representative_digits.shape
```

```python

log_reg= LogisticRegression(max_iter= 10_000)
log_reg.fit(X_representative_digits, y_representative_digits)
log_reg.score(X_test, y_test)
```

Look we jumped from 75% to 83% by just using representative clustered centred digits

Now, check this out

```python
kmeans.labels_
```

```python
y_train_propagated= np.empty(len(X_train), dtype= np.int64)
for i in range(k):
    y_train_propagated[kmeans.labels_ == i] = y_representative_digits[i]
```

```python
log_reg= LogisticRegression(max_iter= 10_000)
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

Even better.. let's do  one last thing... we'll remove the 1% of instances that arre farthest from their **cluster centers**

```python
X_digits_dist[1]
```

```python
percentile_closest= 99

X_cluster_dist= X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster= (kmeans.labels_ == i)
    cluster_dist= X_cluster_dist[in_cluster]
    cutoff_distance= np.percentile(cluster_dist, percentile_closest)
    above_cutoff= (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1

partially_propagated= (X_cluster_dist != -1)
```

```python
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg= LogisticRegression(max_iter= 10_000)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test, y_test)
```

Not a significant impact.. huh.. got carried away there
